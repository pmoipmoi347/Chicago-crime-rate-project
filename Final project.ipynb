```python
# Connect to the database in sql

crime17 = pd.read_csv('Crimes_-_2017.csv')

db17 = sqlite3.connect('crime2017.sql')
cursor17 = db17.cursor()

# Execute query
cursor17.execute('''SELECT *  FROM crime17;''')

final17 = cursor17.fetchall()

df17 = []
for i in final17:
    df17.append(i)
    

header = ('ID', 'Case Number', 'Date', 'Block', 'IUCR', 'Primary Type', 'Description'
          , 'Location Description', 'Arrest', 'Domestic', 'Beat', 'District', 'Ward', 
          'Community Area', 'FBI Code', 'X coordinate', 'Y coordinate', 
          'Year', 'Updated On', 'Latitude', 'Longitude', 'Location')

data17 = pd.DataFrame(data = df17, columns = header)
```

```python
data17
```

```python
cursor17.execute('''select count("Primary Type"),"Primary Type" from crime17 group by "Primary Type"
order by count("Primary Type")desc''')

final17 = cursor17.fetchall()
header = ('Amount', 'TYPE')
data17 = pd.DataFrame(data = final17,columns = header)
```

```python
data17
```

```python
cursor17.execute('''select count("Community Area"),"Community Area", Latitude , Longitude from crime17 group by "Community Area"
order by count("Community Area")desc''')

final17 = cursor17.fetchall()
header = ('Amount', 'Area','Latitude','Longitude')
data17 = pd.DataFrame(data = final17,columns = header)
```

```python
data17
```
